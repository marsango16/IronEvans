group by the data by the Make  using count function

```python
### your code us here
cars.groupby('Make').count()
cars_make=cars.groupby('Make').count()
cars_make
```

Converting Grams/Mile to Grams/Km

1 Mile = 1.60934 Km

Converting Gallons to Liters

1 Gallon = 3.78541 Liters

What brand has the most cars?

```python
### your code us here
#Convert CO2 Emissions from Grams/Mile to Grams/Km
#1 Mile = 1.60934 Km
cars_make['CO2 Emission Grams/Km'] = cars_make['CO2 Emission Grams/Mile']/1.6093
cars_make
```

<b>show the average CO2_Emission_Grams/Km  by Brand

```python
### your code us here
#cars_make['Make']=rownames(cars_make)
#cars_make2= cars_make.groupby("CO2 Emission Grams/Km").max()

cars_make=cars_make.sort_values(by=["CO2 Emission Grams/Mile"],ascending=False)
cars_make["CO2 Emission Grams/Km"].max()
cars_make
```

```python
cars_make1= cars_make
cars_make1.index.name = 'Make2'
cars_make1.reset_index(inplace=True)
cars_make1
```

<b>show the average CO2_Emission_Grams/Km  by Brand ... sorted

```python
### your code us here
#cars_makeCo2=
cars_make2=cars_make1.groupby('Make').mean()
cars_make2.sort_values(by=["CO2 Emission Grams/Km"],ascending=False)
#cars_make
```
