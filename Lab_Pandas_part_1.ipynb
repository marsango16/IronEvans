### Cleaning and wrangling data

- Some car brand names refer to the same brand. Replace all brand names that contain the word "Dutton" for simply "Dutton". If you find similar examples, clean their names too. Use `loc` with boolean indexing.

- Convert CO2 Emissions from Grams/Mile to Grams/Km

- Create a binary column that solely indicates if the transmission of a car is automatic or manual. Use `pandas.Series.str.startswith` and .

- convert MPG columns to km_per_liter

Note:
<br>Converting Grams/Mile to Grams/Km

1 Mile = 1.60934 Km

Converting Gallons to Liters

1 Gallon = 3.78541 Liters



```python
#Some car brand names refer to the same brand. Replace all brand names that contain the word "Dutton" for simply "Dutton". If you find similar examples, clean their names too. Use loc with boolean indexing.

Dutton =cars_df.loc[cars_df['Make'].str.contains('Dutton')]
Dutton

cars_df.loc[cars_df['Make'].str.contains('Dutton'),"Make"] = 'Dutton'
Dutton2 =cars_df.loc[cars_df['Make'].str.contains('Dutton')]
Dutton2

#cars_df=df
#df
#cars_df['Make']
#cars_df
print(cars_df['Make'].str.contains('Dutton'))
```

```python
#Convert CO2 Emissions from Grams/Mile to Grams/Km
#1 Mile = 1.60934 Km
cars_df['CO2 Emission Grams/Km'] = cars_df['CO2 Emission Grams/Mile']/1.6093
cars_df
```

```python
#Create a binary column that solely indicates if the transmission of a car is automatic or manual. 
#Use pandas.Series.str.startswith
cars_df['Auto_Transmission'] = cars_df['Transmission'].str.startswith('Auto')
cars_df['Manual_Transmission'] = cars_df['Transmission'].str.startswith("Manual")
cars_df
#help ("pandas.Series.str.startswith")

```

```python
#convert MPG columns to km_per_liter
cars_df.rename(columns={'City MPG':'City MPG_km_per_liter'}, inplace=True)
cars_df.rename(columns={'Highway MPG':'Highway MPG_km_per_liter'}, inplace=True)
cars_df.rename(columns={'Combined MPG':'Combined MPG_km_per_liter'}, inplace=True)
cars_df
```
