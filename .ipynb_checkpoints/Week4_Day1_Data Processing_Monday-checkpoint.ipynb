```python
#function suitable for for ordinal variables
def replace(i):
    if i == "employed":
        return "1"
    elif i == "unemployed":
        return "2"
    elif i == "medical leave":
        return "3"
    elif i == "disabled":
        return "4"
    elif i == "retired":
        return "5"
    elif i == "m":
        return "1"
    elif i == "f":
        return "2"
    elif i == "rural":
        return "1"
    elif i == "suburban":
        return "2"
    elif i == "urban":
        return "3"
    elif i == "single":
        return "1"
    elif i == "married":
        return "2"
    elif i == "divorced":
        return "3"
    elif i == "no":
        return "1"
    elif i == "yes":
        return "1"
    elif i == "basic":
        return "1"
    elif i == "extended":
        return "2"
    elif i == "premium":
        return "3"
    elif i == "high school or below":
        return "1"
    elif i == "college":
        return "2"
    elif i == "bachelor":
        return "3"
    elif i == "master":
        return "4"
    elif i == "doctor":
        return "5"
    else:
        return i
    
    
```

```python
from sklearn.preprocessing import StandardScaler
data =  numerical["customer_lifetime_value"]
data = np.array(data).reshape(-1,1)
scaler = StandardScaler()
std= scaler.fit(data)
std= std.transform(data)
std= pd.DataFrame(std)
std.columns = ["customer_lifetime_value"]
std
```

```python
#Check that the result has zero mean, unit variance, and reduced skew using mean(), std(), and a plot of the PDF.
varB = std["customer_lifetime_value"].hist(figsize=(10,15))
print("The new hist, after StandardScaling")
```
